```python
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
```

### Augmentation pipeline

```python
transform = transforms.Compose([
    # Resize
    transforms.Resize((500, 500)),
    # Horizontal Flip
    transforms.RandomHorizontalFlip(p=0.5),
    # Color Jitter
    transforms.ColorJitter(brightness=0.239, contrast=0.1, saturation=0.1, hue=0.1),
    # Random Rotation
    transforms.RandomRotation(degrees=2.39),
    # Gaussian Blur
    transforms.GaussianBlur(kernel_size=5),
])
```

### Usage example

```python
test_image = Image.open("test_image.jpg")
augmented_test_img = transform(test_image)
```

### Visualization

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 8))

# Original image
ax[0].imshow(test_image)
ax[0].set_title("Original Image")
ax[0].axis('off')

# Augmented image
ax[1].imshow(augmented_test_img)
ax[1].set_title("Augmented Image")
ax[1].axis('off')

plt.show()
```

### More Visualization

```python
transform_steps = [
    ("Original", lambda x: x),
    ("Resize", transforms.Resize((500, 500))),
    ("Horizontal Flip", transforms.RandomHorizontalFlip(p=1.0)),
    ("Color Jitter", transforms.ColorJitter(brightness=0.239, contrast=0.1, saturation=0.1, hue=0.1)),
    ("Random Rotation", transforms.RandomRotation(degrees=2.39)),
    ("Gaussian Blur", transforms.GaussianBlur(kernel_size=5))
]

fig, axes = plt.subplots(2, len(transform_steps), figsize=(10, 3 * 2))
for i in range(2):
    img = test_image
    for j, (name, transform) in enumerate(transform_steps):
        img = transform(img)
        axes[i, j].imshow(img)
        axes[i, j].set_title(name if i == 0 else "")
        axes[i, j].axis("off")

plt.tight_layout()
plt.show()
```

### Conclusion:

The augmented image is semantically very similar to the original but differs significantly in terms of pixel values.

This can be useful for expanding the training data and improving model robustness.
